# Exploratory Data Analysis on Sample Store

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SampleSuperstore.csv')
df
```

# Correlation

```python
df_corr = df.corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True,cmap='RdBu')
```

# Covariance

```python
df_cov = df.cov()
df_cov
```

```python
sns.heatmap(df_cov,annot=True,cmap='RdYlBu')
```

# Relation between two variables

```python
sns.lmplot(x='Sales',y='Profit',data=df,hue='Segment')
```

# For every segment where Sales increases profit also increases

```python
sns.lmplot(x='Sales',y='Profit',data=df,hue='Category',col = 'Region',col_wrap=2)
```

# From the south region it is clear that for the category of furniture and technology the profit gradually decreases when sales increases.

```python
sns.lmplot(x='Profit',y='Discount',data=df,col = 'Ship Mode',col_wrap = 2,hue='Category')
```

# When profit gets decreasing automatically discount also gets decreased

# Pair plot

```python
sns.pairplot(df)
```

# Box Plot (or) Outliers Detection

```python
sns.boxplot(y=df['Discount'])
```

```python
sns.factorplot(x='Region',y='Sales',data=df,kind='box')
```

```python
sns.boxplot(x='Ship Mode',y='Profit',data=df)
```

```python
sns.catplot(x='Category',y='Sales',data=df,hue='Segment',col = 'Region',col_wrap = 2,kind='box')
```

# Distributions

```python
df.head()
```

```python
df.nunique()
```

```python
sns.distplot(df['Discount'],rug=True)
plt.title('Distribution of Discount')
plt.legend()
```

```python
df['Region'].unique()
```

```python
reg_1 = df[df['Region'] == 'West']
reg_2 = df[df['Region'] == 'South']
reg_3 = df[df['Region'] == 'Central']
reg_4 = df[df['Region'] == 'East']
```

```python
plt.figure(figsize=(14,8))
sns.distplot(reg_1['Profit'],rug=True,hist=False,label = 'West')
sns.distplot(reg_2['Profit'],rug=True,hist=False,label = 'South')
sns.distplot(reg_3['Profit'],rug=True,hist=False,label = 'Central')
sns.distplot(reg_4['Profit'],rug=True,hist=False,label = 'East')
plt.legend()
```

```python
sns.factorplot(x='Segment',y='Sales',data=df,kind='violin',col = 'Region',col_wrap=2)
```
